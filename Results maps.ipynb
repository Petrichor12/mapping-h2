```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('Results/final_df.csv', index_col=0)
#df.sample(20)
```

```python
x = df.loc[:, 'Total Cost per kg H2']
fig = go.Figure(data=go.Scattergeo(
    lat = df.loc[:, 'Latitude'],
    lon = df.loc[:, 'Longitude'],
    mode = 'markers',
    text = round(x, 2),
    marker = dict(
        color = x,
        size = 2.5,
        colorscale = 'Portland',
        #reversescale = True,
        colorbar_title = 'Eur',
        cmax = x.max(),
        colorbar = dict(
            titleside = "right",
            outlinecolor = "rgba(68, 68, 68, 0)",
            ticks = "outside",
            showticksuffix = "last",
            dtick = 1)
    ),
))

fig.update_layout(
    geo = dict(
        showland = True,
        landcolor = "rgb(212, 212, 212)",
        subunitcolor = "rgb(255, 255, 255)",
        countrycolor = "rgb(255, 255, 255)",
        showlakes = True,
        lakecolor = "rgb(255, 255, 255)",
        showsubunits = True,
        showcountries = True,
        projection = dict(
            type = 'natural earth'
        ),
    ),
    title='Total cost per kg H2 (Eur)'
)
fig.show()
```
